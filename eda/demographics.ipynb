```python
exempts = df[df['Vaccine/Exemption'] == 'Exemption']
```

```python
df21 = exempts[exempts['School Year'] == '2020-21']
states = df21['Geography'].unique()
```

```python
df21
```

```python
religs = df21[df21['Dose'] == 'Non-Medical Exemption']
religs = religs[religs['Geography Type'] == 'States']
len(religs), len(states)
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(20,15))
plt.title('Non-Medical Vaccine Exemptions 2020-21, USA Kindergartens')
ax.barh(np.arange(len(religs)), religs['Number of Exemptions'], align='center')
ax.set_yticks(np.arange(len(religs)), labels=religs['Geography'])
ax.invert_yaxis()
# plt.plot(df21['Number of Exemptions'], df21['Geography'])
fig.savefig('eda_non-med-exempts.png')
plt.show()
```
